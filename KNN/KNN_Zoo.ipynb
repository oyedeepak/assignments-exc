```python
# Importing Libraries 
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
mirror= pd.read_csv('C:\\Users\\oyedeepak\\Downloads\\Assignment\\KNN\\Zoo.csv')
mirror.head(10)
```

```python
#Training and Test data using
from sklearn.model_selection import train_test_split
train, test = train_test_split(mirror, test_size= 0.25) #20 percent of entire data
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNC

nei= KNC(n_neighbors= 3, p= 2)
X_train= train.iloc[:, 1:17]
Y_train= train.iloc[:, 17]
X_test= train.iloc[:, 1:17]
Y_test= train.iloc[:, 17]

nei.fit(X_train, Y_train)
```

```python
train_acc= np.mean(nei.predict(X_train)== Y_train)
print(train_acc)

test_acc= np.mean(nei.predict(X_test)== Y_test)
print(test_acc)
```
