```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
df= pd.read_csv('C:\\Users\\oyedeepak\\Downloads\\Assignment\\Decision Tree\\Company_Data.csv')
df.head()
```

```python
df.loc[df['Sales']<=10, 'Sales']=0
df.loc[df['Sales']>10, 'Sales']=1
```

```python
df.head(8)
```

```python
from sklearn import preprocessing

le= preprocessing.LabelEncoder()
for i in range(0,11):
    df.iloc[:,i]= le.fit_transform(df.iloc[:,i])
df.head()
```

```python
# train test split
df_train, df_test= train_test_split(df, test_size= 0.3, random_state=0)
print('train No ', str(len(df_train)))
print('test No ', str(len(df_test)))


train_x= df_train[['ShelveLoc','Urban','US']]
train_y= df_train['Sales']

test_x= df_test[['ShelveLoc','Urban','US']]
test_y= df_test['Sales']
```

```python
train_x.head()
```

```python
feature_names= ['ShelveLoc','Urban','US']
class_names= [0,1]
```

```python
#training the entropy
tree_classifier= tree.DecisionTreeClassifier(criterion= 'gini')
tree_classifier.fit(train_x, train_y)

prediction= tree_classifier.predict(train_x)
accuracy= accuracy_score(train_y, prediction)

print('Trained tree has an accuracy of ', str(accuracy))

prediction= tree_classifier.predict(test_x)
accuracy= accuracy_score(test_y, prediction)

print('Trained tree has an accuracy of ', str(accuracy))
```
