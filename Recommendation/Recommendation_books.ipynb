```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\oyedeepak\Downloads\Assignment\Recommendation System\book.csv", encoding= 'latin-1')
df.head()
```

```python
ratings = pd.DataFrame(df.groupby('Book.Title')['Book.Rating'].mean())
ratings.head()
```

```python
ratings['number_of_ratings'] = df.groupby('Book.Title')['Book.Rating'].count()
ratings.head()
```

```python
import matplotlib.pyplot as plt
ratings['Book.Rating'].hist(bins=20)
plt.show()
#Most of the books have ratings between 8 and 10
```

```python
ratings.sort_values('number_of_ratings', ascending=False).head(10)
```

```python
ratings[ratings['number_of_ratings'] > 3].sort_values('Book.Rating', ascending=False).head(10) #select books which are rated by 4 or more readers
```
