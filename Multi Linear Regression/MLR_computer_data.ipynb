```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
computer = pd.read_csv("C:/Users/oyedeepak/Downloads/Assignment/Multi Linear Regression/Computer_Data.csv"); computer.head()
```

```python
computer.drop(computer.columns[0], axis =1, inplace = True)
```

```python
computer.corr()
```

```python
computer.dtypes
```

```python
CD = pd.get_dummies(computer.cd, prefix=' CD').iloc[:,1]
Multi = pd.get_dummies(computer.multi, prefix='Multi').iloc[:,1]
Premium = pd.get_dummies(computer.premium, prefix='Premium').iloc[:,1]
```

```python
merged= pd.concat([computer, CD, Multi, Premium], axis = 1)
merged= merged.drop(['cd','multi','premium'], axis= 1)
merged
```

```python
X= merged.drop('price', axis= 1)
Y= merged.price
```

```python
# splitting the dataset into test and train

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size= 0.2, random_state= 5)
```

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
# predicting the Test set results

Y_pred= regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
score = r2_score(Y_test, Y_pred)
```
