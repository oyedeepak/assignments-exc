```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

from mlxtend.preprocessing import TransactionEncoder

from mlxtend.frequent_patterns import apriori, association_rules 
```

```python
df = pd.read_excel(r"C:\Users\oyedeepak\Downloads\Assignment\Association Rules\groceries.xlsx", header= None)
df.head(10)
```

```python
#converting dataframe to array
df1 = df.to_numpy()
df1
```

```python
#removing 'NaN'
df2 = [[i for i in j if i == i] for j in df1]
df2
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()

dfn = te.fit(df2).transform(df2)
dfn
```

```python
dfn.astype("int")
```

```python
te.columns_
```

```python
df3 = pd.DataFrame(dfn, columns=te.columns_)
```

```python
df3
```

```python
#f = lambda x: 1 if x==True else 0
#df3.applymap(f)
#df3.head(20)
```

```python
frequent_itemsets = apriori(df3, min_support=0.005, max_len = 3, use_colnames=True)
print (frequent_itemsets.head(20))
```

```python
# Most Frequent item sets based on support 
frequent_itemsets.sort_values('support',ascending = False, inplace=True)
plt.barh(list(range(1,11)),frequent_itemsets.support[1:11],color='bgcmyr')
plt.yticks(list(range(1,11)),frequent_itemsets.itemsets[1:11])
plt.ylabel('Item-Support')
plt.xlabel('support')
plt.show()
```

```python
#association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules.antecedents = rules.antecedents.apply(lambda x: next(iter(x)))
rules.consequents = rules.consequents.apply(lambda x: next(iter(x)))
rules
```

```python
import networkx as nx


fig, ax=plt.subplots(figsize=(10,8))
GA=nx.from_pandas_edgelist(rules,source='antecedents',target='consequents')
nx.draw(GA,with_labels=True)
plt.show()
```
